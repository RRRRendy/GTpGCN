```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

# Load the data from the uploaded CSV file
file_path = '/home/rendy/fMRI/DATASETS/MDD/848MDDvs794NC/848MDDvs794NC_CC200.csv'
data = pd.read_csv(file_path)

# Setting the aesthetic style of the plots
sns.set(style="ticks")  # Use 'ticks' style which doesn't have gridlines

# Increase the font scale for better readability in publications
sns.set_context("paper", font_scale=2.5)

# Setting font to Arial
rcParams['font.family'] = 'Arial'

# Filtering data for MDD and HC
mdd_data = data[data['Dx'] == 1]
hc_data = data[data['Dx'] == 0]

def plot_site_data(site_data, title_prefix):
    # Create figure with two subplots
    fig, axes = plt.subplots(1, 2, figsize=(16, 7))  # Adjusted for better spacing and size

    # Plot for age distribution
    sns.boxplot(x='Site', y='Age', data=site_data, ax=axes[0])
    axes[0].set_title(f'{title_prefix} - Age Distribution by Site', fontsize=32)
    axes[0].set_xlabel('Site')
    axes[0].set_ylabel('Age')
    axes[0].tick_params(axis='x', labelrotation=45)  # Rotate x labels for better visibility
    axes[0].tick_params(labelsize=18)  # Increase tick label font size

    # Plot for sex distribution
    sex_counts = site_data.groupby(['Site', 'Sex']).size().unstack(fill_value=0)
    sex_counts.plot(kind='bar', stacked=True, ax=axes[1], color=['#4C72B0', '#DD8452'])
    axes[1].set_title(f'{title_prefix} - Sex Distribution by Site', fontsize=32)
    axes[1].set_xlabel('Site')
    axes[1].set_ylabel('Count')
    axes[1].legend(title='Sex', labels=['Female', 'Male'])
    axes[1].tick_params(axis='x', labelrotation=45)  # Rotate x labels for better visibility
    axes[1].tick_params(labelsize=18)  # Increase tick label font size

    plt.tight_layout()
    plt.show()

# Plotting for MDD and HC separately
plot_site_data(mdd_data, 'MDD')
plot_site_data(hc_data, 'HC')
```
